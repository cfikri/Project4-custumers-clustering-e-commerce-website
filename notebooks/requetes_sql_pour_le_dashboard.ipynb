### 1. En excluant les commandes annulées, quelles sont les commandes récentes de moins de 3 mois que les clients ont reçu avec au moins 3 jours de retard ?

```python
%%sql

SELECT *
FROM orders
WHERE order_status = 'delivered'
      AND order_purchase_timestamp > DATE((SELECT max(order_purchase_timestamp) FROM orders), '-3 months')
      AND order_delivered_customer_date >= DATE(order_estimated_delivery_date, '+3 days');
```

### 2. Qui sont les vendeurs ayant généré un chiffre d'affaires de plus de 100 000 Real sur des commandes livrées via Olist ?

```python
%%sql

SELECT seller_id, sum(order_item_id*price) AS chiffre_affaires
FROM order_items i
JOIN orders o ON i.order_id = o.order_id
WHERE order_status = 'delivered'
GROUP BY seller_id
HAVING chiffre_affaires > 100000;
```

### 3. Qui sont les nouveaux vendeurs (moins de 3 mois d'ancienneté) qui sont déjà très engagés avec la plateforme (ayant déjà vendu plus de 30 produits) ?

```python
%%sql

SELECT seller_id, sum(order_item_id) AS nombre_ventes, min(order_purchase_timestamp) AS anciennete
FROM order_items i
JOIN orders o ON i.order_id = o.order_id 
WHERE order_status = 'delivered'
GROUP BY seller_id
HAVING nombre_ventes > 30
       AND anciennete > DATE((SELECT max(order_purchase_timestamp) FROM orders), '-3 months');
```

### 4. Quels sont les 5 codes postaux, enregistrant plus de 30 commandes, avec le pire review score moyen sur les 12 derniers mois ?

```python
%%sql

SELECT seller_zip_code_prefix AS code_postal, sum(order_item_id) AS nombre_commandes, avg(review_score) AS score_moyen
FROM order_items i
JOIN sellers s ON i.seller_id = s.seller_id
JOIN order_reviews r ON i.order_id = r.order_id
WHERE review_answer_timestamp >= DATE((SELECT max(review_answer_timestamp) FROM order_reviews), '-12 months')
GROUP BY code_postal
HAVING nombre_commandes > 30
ORDER BY score_moyen LIMIT 5;
```
